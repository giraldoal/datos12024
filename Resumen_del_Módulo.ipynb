# Resumen del módulo
En el módulo se realizó un estudio basando en ejemplos acerca del uso de Python para el análisis de datos, particularmente en la extracción, transformación y carga de datos (ETL) de diferentes fuentes como archivos PDF, Excel y de texto plano.

## Extracción

* Archivos PDF: Se utiliza la librería PDFQuery para extraer información específica de archivos PDF. Se definen selectores para ubicar la información dentro del documento utilizando la estructura del PDF y el texto que lo compone.
* Archivos Excel: Se emplea la librería Openpyxl para leer y extraer datos de archivos Excel. Se accede a la información por hojas, filas y columnas específicas.
* Archivos de texto plano: Se utiliza el módulo os para abrir y leer archivos de texto, y el módulo re con expresiones regulares para buscar patrones y extraer información específica.


## Transformación
* Limpieza: Los datos extraídos a menudo requieren limpieza, como eliminar espacios en blanco, convertir tipos de datos o manejar valores faltantes.
* Estructuración: Se organizan los datos en estructuras de datos de Python como listas, diccionarios o DataFrames de Pandas.
* Agregación y cálculo: Se pueden realizar operaciones de agregación, como calcular promedios, contar ocurrencias o agrupar datos. También se pueden calcular nuevos indicadores a partir de los datos existentes.

## Carga
* Archivo CSV: El formato de archivo más común para cargar los datos transformados es el CSV (valores separados por comas). Se utiliza la función to_csv de Pandas para guardar los DataFrames en archivos CSV.

## Ejemplos de aplicación
* Análisis de expresión genética: Se utiliza Numpy para analizar una matriz de expresión genética, seleccionando subconjuntos de genes y tejidos.
* Recolección automatizada de frutas: Se procesan imágenes de manzanas para identificar las maduras basándose en el color. Se compara la eficiencia del procesamiento usando listas de Python y arrays de Numpy.
* Análisis del presupuesto del SGP: Se extrae información de varios archivos Excel para analizar la distribución del presupuesto del Sistema General de Participaciones en diferentes sectores y municipios.
* Análisis de datos de vinos: Se utiliza Pandas para analizar un conjunto de datos de vinos, incluyendo el conteo de clases, la separación de datos por clase y el cálculo de estadísticas descriptivas.
* Caracterización de zonas PDET: Se extrae información de fichas PDF para caracterizar las zonas PDET en Colombia, incluyendo datos de población, pobreza y empleo.
