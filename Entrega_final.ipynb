```python
# Eliminar las filas que contienen NaN en la columna 'Etnia'
df_deplim = df_dep.dropna(subset=['Etnia'])

# Mostrar las primeras filas del DataFrame después de eliminar NaN en 'Etnia'
df_deplim.head()
```

```python
# Contar la frecuencia de cada valor en Etnia
frecuencia = df_deplim['Etnia'].value_counts()

# Calcular el porcentaje de cada valor
porcentaje = df_deplim['Etnia'].value_counts(normalize=True) * 100

# Combinar frecuencia y porcentaje en un DataFrame
tabla_etnia = pd.DataFrame({
    'Frecuencia': frecuencia,
    'Porcentaje (%)': porcentaje
})
print(tabla_etnia)
```

```python
# Reemplazar todos los valores de 'Etnia' que no sean 'Negro' por 'Otro'
df_deplim['Etnia'] = df_deplim['Etnia'].apply(lambda x: 'Otro' if x != 'Negro' else x)

# Verificar la frecuencia de los valores después de agrupar
frecuencia = df_deplim['Etnia'].value_counts()

# Calcular el porcentaje de cada valor después de agrupar
porcentaje = df_deplim['Etnia'].value_counts(normalize=True) * 100

# Combinar frecuencia y porcentaje en un DataFrame
tabla_etnia = pd.DataFrame({
    'Frecuencia': frecuencia,
    'Porcentaje (%)': porcentaje
})

# Mostrar la tabla resultante
print(tabla_etnia)
```

```python

# Definir las variables categóricas y numéricas
categorical_features = ['Sexo', 'Etnia', 'HTA', 'DM', 'IECA', 'ARA2', 'Estado_Renal']
numeric_features = ['Edad', 'IMC', 'TAS']

# Crear un Pipeline para las variables numéricas (imputación y estandarización)
numeric_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='mean')),  # Reemplazar NaN en numéricas con la media
    ('scaler', StandardScaler())                 # Escalar las variables numéricas
])

# Crear un Pipeline para las variables categóricas (imputación con moda)
categorical_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent'))  # Reemplazar NaN en categóricas con la moda
])

# Crear un ColumnTransformer que aplique las transformaciones a las variables categóricas y numéricas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', numeric_transformer, numeric_features),   # Aplicar a las variables numéricas
        ('cat', categorical_transformer, categorical_features)  # Aplicar a las variables categóricas
    ]
)

# Ajustar el preprocesador a los datos y transformar el DataFrame
df_cleaned = preprocessor.fit_transform(df_deplim)

# Obtener los nombres de las columnas categóricas y numéricas procesadas
column_names = numeric_features + categorical_features

# Convertir los datos transformados en un DataFrame
df_transformed = pd.DataFrame(df_cleaned, columns=column_names)

# Mostrar las primeras filas del DataFrame con los valores imputados
print(df_transformed.head())
```

```python
from statsmodels.stats.power import ztest_power

# Definir los parámetros
p1 = 0.60  # Proporción del grupo con intervención convencional
p2 = 0.40  # Proporción del grupo con la nueva intervención
alpha = 0.05  # Nivel de significancia (5%)
power = 0.80  # Poder del 80%

# Diferencia de proporciones
effect_size = abs(p1 - p2)

# Cálculo del tamaño de la muestra
sample_size = ztest_power(effect_size=effect_size, nobs1=None, alpha=alpha, power=power, alternative='two-sided')

# Multiplicamos por 2 para tener el tamaño de muestra total (dos grupos)
sample_size_total = sample_size * 2

# Mostrar el tamaño de la muestra
print(f"Tamaño de muestra total: {round(sample_size_total)}")
```
